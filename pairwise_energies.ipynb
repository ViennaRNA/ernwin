```python
import collections as col

def load_loops(filename):
    f = open(filename, 'r')
    fractions = col.defaultdict(list)
    sum = 0
    for line in f:
        parts = line.strip().split()
        if len(parts) == 2:
            fractions[int(parts[0])] += [float(parts[1])]
            sum += 1
    return fractions

loops_native = load_loops('fess/stats/loop_loop3_distances_native.csv')
loops_sampled = load_loops('fess/stats/loop_loop3_distances_sampled.csv')
```

```python
hist(loops_native)
hist(loops_sampled)
```

```python
def load_metaloop_energies(fn):
    lengths = []
    energies = []
    with open(fn, 'r') as f:
        for line in f:
            parts = line.strip().split()
            x = parts[1].strip('[]')
            
            if len(x) == 0:
                continue
                
            lengths += [int(parts[0])]
            energies += [float(parts[1].strip('[]'))]
            
    return (lengths, energies)

(lengths_native, energies_native) = load_metaloop_energies('temp/loop_loop4_native.csv')
(lengths_sampled, energies_sampled) = load_metaloop_energies('temp/loop_loop4_sampled.csv')
            
```

```python
fig, ax = plt.subplots()

native = np.array([lengths_native, energies_native]).T
sampled =  np.array([lengths_sampled, energies_sampled]).T

ax.set_xlim([0,5])

ax.plot(lengths_native, energies_native, 'o')
ax.plot(lengths_sampled, energies_sampled, 'o')

```
