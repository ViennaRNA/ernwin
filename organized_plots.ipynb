```python
import fess.builder.energy as fbe
import brewer2mpl
import itertools as it

def check_energy_distribution((energy, struct_name), real_distribution_fn = None, 
                              sampled_distribution_fn = None, xlim=None, ax=None, plot_hists = True, 
                              ylim=None, energy_color=None, energy_label=None, legend=True, filename=None,
                              xlab=None, ylab=None, ylab2=None, legend_outside=False, title=None, figsize=None, grid=False,
                              draw_actual_value=True, show_density=True):
    '''
    Visually display the probability distribution represented by
    the energy passed in as the first parameter.

    @param energy: The energy to be checked
    @param real_distribution_fn: The filename of the distribution of real values
    @param sampled_distribution_fn: The filename of the distribution of sampled values
    @param struct_name: The structure to test it for
    '''
    target_struct = structs[struct_name].bg
    energy.eval_energy(structs[struct_name])
    
    # The energy function should know how to parse its own values file
    if real_distribution_fn is not None:
        energy.real_stats_fn = real_distribution_fn
        (kr, vals) = energy.get_distribution_from_file(real_distribution_fn, energy.measure_category(target_struct))
    else:
        kr = energy.real_kdes[energy.measure_category(target_struct)]
    
    if sampled_distribution_fn is not None:
        energy.sampled_stats_fn = sampled_distribution_fn
        (ks, vals) = energy.get_distribution_from_file(sampled_distribution_fn, energy.measure_category(target_struct))
    else:
        ks = energy.sampled_kdes[energy.measure_category(target_struct)]
    
    if xlim is None:
        xlim = (min(it.chain(energy.vals[0], energy.vals[1])),
                0.85 * max(it.chain(energy.vals[0], energy.vals[1])))
        
    if xlim is None:
        xs = np.linspace(energy.flocs[0], energy.fscales[0], 100)
    else:
        xs = np.linspace(xlim[0], xlim[1], 100)
    
    if ax is None:
        if figsize is None:
            fig, ax1 = plt.subplots()
        else:
            fig = plt.figure(figsize=figsize)
            ax1 = fig.add_subplot(1,1,1)
    else:
        ax1 = ax
    
    dark2 = brewer2mpl.get_map('Set1', 'qualitative', 4).mpl_colors[2:4]
    
    if plot_hists:
        h1 = ax1.hist(energy.vals[0], normed=True, alpha=0.6, label='Target Measures', bins=num_bins, color=dark2[0])
        h2 = ax1.hist(energy.vals[1], normed=True, alpha=0.6, label='Reference Measures', bins=num_bins, color=dark2[1])
        
        p1 = ax1.plot(xs, kr(xs), ls='--', label='Target Fit', color=dark2[0])
        p2 = ax1.plot(xs, ks(xs), ls='--', label = 'Reference Fit', color=dark2[1])
        
    if xlab is None:
        ax1.set_xlabel('Coarse Grain Measure', fontsize=axislabel_fontsize)
    else:
        ax1.set_xlabel(xlab, fontsize=axislabel_fontsize)
        
    if ylab is None:
        ax1.set_ylabel('Density', fontsize=axislabel_fontsize)
    else:
        ax1.set_ylabel(ylab, fontsize=axislabel_fontsize)
    
    if title is None:
         plt.text(0.5, 1.06, struct_name,
         horizontalalignment='center',
         fontsize=title_fontsize,
         transform = ax1.transAxes)
    else:
         plt.text(0.5, 1.06, energy.get_name(),
         horizontalalignment='center',
         fontsize=title_fontsize,
         transform = ax1.transAxes)
    
    vals_real = energy.vals[0]
    vals_sampled = energy.vals[1]
    
    #plot_real_vs_sampled(vals_real, vals_sampled, floc=energy.flocs[0], xlim=[energy.flocs[0], energy.fscales[0]])
    actual_value = energy.get_cg_measure(structs[struct_name])
    #fud.pv('actual_value')
    #ax1.axvline(actual_value, color='k', ls=':', lw=6 )
    if draw_actual_value:
        ax1.axvline(actual_value)
    
    ax2 = ax1.twinx()
    
    if energy_label is None:
        energy_label = 'Energy'
    
    energies = -(np.log(kr(xs) + 0.00001 * ks(xs)) - np.log(ks(xs)))
    if energy_color is None:
        p3 = ax2.plot(xs, energies, label=energy_label)
    else:
        p3 = ax2.plot(xs, energies, label=energy_label, color=energy_color)
        
    #ax2.set_ylabel('Energy Value')
    
    ax1.set_xlim(xlim)
    ax2.set_xlim(xlim)
    
    if ylim is not None:
        ax2.set_ylim(ylim)
    #plot(xs, ysr)
    #plot(xs, yss)

    ls1, lb1 = ax1.get_legend_handles_labels()
    ls2, lb2 = ax2.get_legend_handles_labels()
    
    ax1.grid(grid)
    ax2.grid(grid)
    
    if ylab2 is not None:
        ax2.set_ylabel(ylab2, fontsize=axislabel_fontsize)
        
    if not show_density:
        ax1.get_yaxis().set_visible(False)
        
    if legend:
        if legend_outside:
            plt.legend(ls1 + ls2, lb1 + lb2, bbox_to_anchor=(1.20, 0.9), loc=2, borderaxespad=0.)
        else:
            plt.legend(ls1 + ls2, lb1 + lb2, loc=1, borderaxespad=0.)
        
    if filename is not None:
        plt.savefig(filename, bbox_inches='tight')
    
    '''
    print "energy: %.3f measure: %.3f" % (energy.eval_energy(structs[struct_name]), 
                                          energy.get_cg_measure(structs[struct_name]))
    '''
    
    return (ls1 + ls2, lb1 + lb2)

check_energy_distribution((fbe.ShortestLoopDistanceEnergy(), '1y26'), legend_outside=True, 
filename='1y26_shortestloopdistanceenergy.pdf', title='Loop Loop Energy', figsize=(6,2.5))
pass
```

```python
def plot_multiple(plot_function, args, filename=None, figsize=(14,9), label_left = "Density", 
                  label_bottom="Coarse Grain Measure Value", label_right="Energy", legend_pos=(1.40,0.90)):
    
    fig = plt.figure(figsize=figsize)
    #figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    num_structs = len(args)
    rows = int(1.5 * math.sqrt(num_structs))
    cols = int(math.ceil(num_structs / float(rows)))
    
    #figsize(cols * 4, rows * 3)
    
    for i, arg in enumerate(args):
        ax = fig.add_subplot(rows, cols, i+1)
        (ls, lb) = plot_function(arg, ax=ax)
            
    ax1 = fig.add_subplot(rows, cols, 1)
    ax = fig.add_subplot(rows, cols, num_structs)
    #ls1, lb1 = ax1.get_legend_handles_labels()
    
    if label_bottom is not None:
        fig.text(0.5, 0.00, label_bottom, ha='center', va='center', fontsize=13)
        
    if label_left is not None:
        fig.text(0.00, 0.60, label_left, ha='center', va='center', rotation='vertical', fontsize=13)
    
    if label_right is not None:
        fig.text(1., 0.60, label_right, ha='center', va='center', rotation='vertical', fontsize=13)
    
    plt.tight_layout()
    plt.subplots_adjust(top=1.30)
    
    plt.legend(ls, lb, bbox_to_anchor=legend_pos, loc=2, borderaxespad=0.)
    
    if filename is not None:
        savefig(filename, bbox_inches='tight')
```

```python
import re
import math

import scipy.stats as ss
import brewer2mpl

def get_rmsds(log_file):
    '''
    Plot the rmsds of the sampled structures in the given log file.
    '''
    rmsds = []
    
    with open(log_file, 'r') as f:
        for line in f:
            if line.find('native_energy') == 0:
                parts = re.split(' +', line)
                rmsd = float(parts[5])
                
                rmsds += [rmsd]
    return rmsds

def plot_rmsds_core(paths, ax = None, color=None, xlab=None, ylab=None, grid=False):
    if ax is None:
        fig, ax = plt.subplots()

    if len(paths) < 3:
        dark2 = brewer2mpl.get_map('Dark2', 'qualitative', 3).mpl_colors[:2][::-1]
    else:
        dark2 = brewer2mpl.get_map('Dark2', 'qualitative', len(paths)).mpl_colors
        
    for i, path in enumerate(paths):
        rmsds = get_rmsds(op.join(path, 'log.txt'))
        xs = np.linspace(min(rmsds), max(rmsds))
        k = ss.gaussian_kde(rmsds)
        kxs = k(xs)
        
        densest_rmsd = xs[kxs == max(kxs)]
        #densest_rmsd = sum(xs * kxs)
        #print densest_rmsd
        

        npath = op.normpath(path)
        struct = op.basename(npath)
        category = op.basename(op.dirname(npath))
        
        ax.hist(rmsds, alpha=0.3, normed=True, color=dark2[i], bins=num_bins)
        ax.plot(xs, k(xs), label=category, color=dark2[i])
        
        ax.axvline(densest_rmsd, color=dark2[i], ls='--')
    
    if xlab is not None:
        ax.set_xlabel(xlab)
    if ylab is not None:
        ax.set_ylabel(ylab)
        
    ax.set_title(struct)
    ax.set_xlim(0, max(rmsds))
    ax.grid(grid)
    ls, lb = ax.get_legend_handles_labels()
    
    return (ls, lb)
```

```python
import scipy.stats as ss
import copy
import brewer2mpl

def plot_target_and_sampled_cg((energy, directory, constant), ax=None, 
                               filename=None, title=None, legend=False, figsize=None, xlab=None,
                               ylab=None, grid=False):
    '''
    Plot the target coarse grain distribution as well as all of the values
    sampled
    '''
    energy = copy.deepcopy(energy)
    fn = op.basename(directory)
    real_cg = ftmc.CoarseGrainRNA(op.join(op.expanduser('~/doarse'), fn, 'temp.cg'))

    cg = ftmc.CoarseGrainRNA(op.join(directory, 'best0.coord'))
    energy.eval_energy(fbm.SpatialModel(cg))

    sampled_vals = np.genfromtxt(op.join(directory, energy.get_energy_name()), delimiter=' ')
    kde_sampled = ss.gaussian_kde(sampled_vals)
    
    constant_vals = np.genfromtxt(op.join(constant, energy.get_energy_name()), delimiter=' ')
    kde_constant = ss.gaussian_kde(constant_vals)
    
    real_cg_measure = energy.get_cg_measure(fbm.SpatialModel(real_cg))
    #print real_cg_measure
    
    mn = min(it.chain(energy.vals[0], constant_vals, sampled_vals))
    mx = max(it.chain(energy.vals[0], constant_vals, sampled_vals))
    
    xs = np.linspace(mn, mx, 100)
    
    rk = energy.real_kdes[energy.measure_category(cg)]
    if ax is None:
        if figsize is not None:
            fig = plt.figure(figsize=figsize)
            ax = fig.add_subplot(1,1,1)
        else:
            fig, ax = plt.subplots()
    
    dark2 = brewer2mpl.get_map('Dark2', 'qualitative', 3).mpl_colors
    
    ax.hist(constant_vals, normed=True, alpha=0.4, color=dark2[0], bins=num_bins)
    ax.hist(energy.vals[0], normed=True, alpha=0.4, color=dark2[1], bins=num_bins)
    ax.hist(sampled_vals, normed=True, alpha=0.4, color=dark2[2], bins=num_bins)

    ax.plot(xs, kde_constant(xs), color=dark2[0], label="Background")
    ax.plot(xs, rk(xs), color=dark2[1], label="Target")
    ax.plot(xs, kde_sampled(xs), color=dark2[2], label="Sampled")
    ax.axvline(real_cg_measure)
    ax.grid(grid)
    
    if xlab is None:
        ax.set_xlabel('Coarse Grain Measure', fontsize=axislabel_fontsize)
    else:
        ax.set_xlabel(xlab, fontsize=axislabel_fontsize)
        
    if ylab is None:
        ax.set_ylabel('Density', fontsize=axislabel_fontsize)
    else:
        ax.set_ylabel(ylab, fontsize=axislabel_fontsize)
    
    if title is None:
        
        ax.set_title(fn)
    else:
        plt.text(0.5, 1.06, title,
         horizontalalignment='center',
         fontsize=title_fontsize,
         transform = ax.transAxes)
    
    ls1, lb1 = ax.get_legend_handles_labels()
    if legend:
        plt.legend(ls1, lb1, bbox_to_anchor=(1.05, 0.9), loc=2, borderaxespad=0.)
        
    if filename is not None:
        savefig(filename, bbox_inches='tight')
    
    return (ls1, lb1)

#plot_target_and_sampled_cg(fbe.RadiusOfGyrationEnergy(), 'new_sampled/discrete/rog/1il2', 'new_sampled/discrete/constant-energy/1il2')
```

```python
def plot_energies(tnew, filename=None, use_ernwin_name=False):
    tnewer = tnew.pivot(index='energy', columns='pdb')
    m = tnewer.as_matrix()
    tnewer.columns = tnewer.columns.droplevel(0)
    
    sizes = dict()
    sizelist = []
    for p in tnewer.columns:
        sizes[p] = structs[p].bg.seq_length
        sizelist += [structs[p].bg.seq_length]
    
    xargsrt = sorted(range(len(sizelist)), key=sizelist.__getitem__)
    #xargsrt=np.mean(m.T, axis=1).argsort()
    m = m.T[xargsrt].T
    yargsrt = np.mean(m, axis=1).argsort()
    m = m[yargsrt]
    
    fig, ax = plt.subplots()
    im = ax.imshow(-m, interpolation='none', cmap=plt.get_cmap('RdYlGn'))
    
    cols = tnewer.index[yargsrt]
    newcols = []
    
    for col in cols:
        if use_ernwin_name:
            print col
            if col == 'cylinder-shortestloop-rog':
                col = 'ernwin'
        newcols += [col]
    
    ax.set_xticks(range(len(tnewer.columns)))
    ax.set_xticklabels([c + " [%d]" % (sizes[c]) for c in tnewer.columns[xargsrt]], rotation=90)
    ax.set_yticks(range(len(tnewer.index)))
    ax.set_yticklabels(newcols)
    
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    
    cb = plt.colorbar(im, cax=cax)

    cb.set_label('-RMSD')
    
    if filename is not None:
        plt.savefig(filename, bbox_inches='tight')
    
def filter_energies(t, measure_type, stats_type):
    ts = t[t[0] == measure_type]
    ts = ts[ts[5] == stats_type]

    #print ts.head()
    tnew = pa.DataFrame(data={"energy": ts[6], "pdb":ts[7], "rmsd":ts[3]})
    return tnew

```
